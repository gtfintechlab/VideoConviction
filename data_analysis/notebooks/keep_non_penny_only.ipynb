```python
import pandas as pd
from polygon import RESTClient
```

```python
df = pd.read_csv("../data_source/complete_dataset.csv")
df = df[df['is_rec_present'] == 'Yes']
print("Number of recommendations in total: ", len(df))
```

```python
ticker_corrections = {
    'APPL': 'AAPL', 
    'GOOGL': 'GOOG', 
    'FB': 'META',
    'GLDLF': 'GLDG',
    'HLIX': 'HLX',
    'AI†': 'AI',
    'SKWS': 'SWKS'
}

# Clean and correct ticker names
df['ticker_name'] = df['ticker_name'].str.replace(r'^[A-Z]+:\s*', '', regex=True)  # Remove "NYSE: ", etc.
df['ticker_name'] = df['ticker_name'].replace(ticker_corrections)  # Apply corrections
```

```python
# Replace 'YOUR_API_KEY' with your actual Polygon.io API key
api_key = 'PFmF97gf4Lx4mCYqainUhBkosphzL59h'
client = RESTClient(api_key=api_key)
```

```python
def get_current_price(ticker):
    try:
        last_trade = client.get_last_trade(ticker)
        return last_trade.price
    except Exception as e:
        print(f"Error fetching price for {ticker}: {e}")
        return None

# Add current price column
df["current_price"] = df["ticker_name"].apply(get_current_price)
```

```python
print("Number of recommendations in total: ", len(df))
# FILTER
df = df[df['current_price'] > 5]
print("Number of NON PENNY recommendations: ", len(df))
```
